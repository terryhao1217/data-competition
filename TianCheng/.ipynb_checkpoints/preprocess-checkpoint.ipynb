```python
#下一步从特征重要性，组合特征，数量特征最值，均值角度考虑
import pandas as pd
```

```python
operation_train=pd.read_csv('data/operation_train_new.csv')
transaction_train=pd.read_csv('data/transaction_train_new.csv')
tag_train=pd.read_csv('data/tag_train_new.csv')
```

```python
operation_test=pd.read_csv('data/operation_round1_new.csv')
transaction_test=pd.read_csv('data/transaction_round1_new.csv')
```

```python
#训练集数量频词特征
operation_train_feature=operation_train.groupby('UID').size().reset_index()
del operation_train_feature[0]
operation_train_feature['operation_total_num']=operation_train.groupby('UID').size().reset_index()[0]
operation_train_feature['operation_day_num']=operation_train.groupby('UID')['day'].nunique().reset_index()['day']

transaction_train_feature=transaction_train.groupby('UID').size().reset_index()
del transaction_train_feature[0]
transaction_train_feature['transaction_total_num']=transaction_train.groupby('UID').size().reset_index()[0]
transaction_train_feature['transaction_day_num']=transaction_train.groupby('UID')['day'].nunique().reset_index()['day']
```

```python
#测试集数量频词特征
operation_test_feature=operation_test.groupby('UID').size().reset_index()
del operation_test_feature[0]
operation_test_feature['operation_total_num']=operation_test.groupby('UID').size().reset_index()[0]
operation_test_feature['operation_day_num']=operation_test.groupby('UID')['day'].nunique().reset_index()['day']

transaction_test_feature=transaction_test.groupby('UID').size().reset_index()
del transaction_test_feature[0]
transaction_test_feature['transaction_total_num']=transaction_test.groupby('UID').size().reset_index()[0]
transaction_test_feature['transaction_day_num']=transaction_test.groupby('UID')['day'].nunique().reset_index()['day']
```

```python
operation_test_by_day=operation_test.groupby(('UID','day')).size()
for i in range(1,31):
    feature='operation_day_'+str(i)+'_num'
    operation_test_feature[feature]=0
    for index0 in operation_test_feature['UID']:
        if  i in operation_test_by_day[index0].index:
            operation_test_feature.loc[operation_test_feature['UID']==index0,feature]=operation_test_by_day[index0][i]
        else:
            operation_test_feature.loc[operation_test_feature['UID']==index0,feature]=0
    print('finish the '+str(i)+'th epoch') 
operation_test_feature.to_csv('data/operation_test_feature.csv')

transaction_test_by_day=transaction_test.groupby(('UID','day')).size()
for i in range(1,31):
    for index0 in transaction_test_feature['UID']:
        feature='transaction_day_'+str(i)+'_num'
        transaction_test_feature[feature]=0
        if  i in transaction_test_by_day[index0].index:
            transaction_test_feature.loc[transaction_test_feature['UID']==index0,feature]=transaction_test_by_day[index0][i]
           
        else:
            transaction_test_feature.loc[transaction_test_feature['UID']==index0,feature]=0
    print('finish the '+str(i)+'th epoch')
transaction_test_feature.to_csv('data/transaction_test_feature.csv')
```
