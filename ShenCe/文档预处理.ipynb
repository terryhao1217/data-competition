```python
import pandas as pd
import re
import jieba
import jieba.analyse
import jieba.posseg as pseg
```

```python
raw_data=pd.read_csv('data/all_docs.txt',sep='\001',header=None)
raw_data.columns=['id','title','doc']


train_data=pd.read_csv('data/train_docs_keywords.txt',sep='\t')
train_data.columns=['id','label']
train_id_list=list(train_data['id'].unique())
train_title_doc=raw_data[raw_data['id'].isin(train_id_list)]

test_title_doc=raw_data[~raw_data['id'].isin(train_id_list)]
```

```python
def cut_byPos(sentence):
    words_list=[]
    pos_list=['ns','s','nz','nr','vn','m','v','n','an']
    cut_sentence=pseg.cut(sentence)
    for word,flag in cut_sentence:
        if flag in pos_list:
            words_list.append(word)  

    print('finish i articles')

    return words_list
    
```

```python

train_title_doc['title']=train_title_doc['title'].astype(str)
train_title_doc['title']=train_title_doc['title'].apply(lambda x:''.join(re.split(r"《.+?》",x)))
train_title_doc['title_cut']=train_title_doc['title'].apply(lambda x:','.join(jieba.analyse.extract_tags(x,topK=5,allowPOS=('ns','nz','nr','vn','v','n','an'))))

train_title_doc['doc']=train_title_doc['doc'].astype(str)
train_title_doc['doc']=train_title_doc['doc'].apply(lambda x:''.join(re.split(r"《.+?》",x)))
train_title_doc['doc']=train_title_doc['doc'].apply(lambda x:''.join(filter(lambda ch:ch not in ' \t1234567890，？。！：',x)))
i=0
train_title_doc['doc_cut']=train_title_doc['doc'].apply(lambda x:','.join(cut_byPos(x)))

train_title_doc['article_cut']=train_title_doc['title_cut']+train_title_doc['doc_cut']
train_title_doc['article_cut'].to_csv('data/fenci_doc_train.txt',index=None,header=None)
```

```python

test_title_doc['title']=test_title_doc['title'].astype(str)
test_title_doc['title']=test_title_doc['title'].apply(lambda x:''.join(re.split(r"《.+?》",x)))
test_title_doc['title_cut']=test_title_doc['title'].apply(lambda x:','.join(jieba.analyse.extract_tags(x,topK=5,allowPOS=('ns','nz','nr','vn','v','n','an'))))

test_title_doc['doc']=test_title_doc['doc'].astype(str)
test_title_doc['doc']=test_title_doc['doc'].apply(lambda x:''.join(re.split(r"《.+?》",x)))
test_title_doc['doc']=test_title_doc['doc'].apply(lambda x:''.join(filter(lambda ch:ch not in ' \t1234567890，？。！：',x)))
i=0
test_title_doc['doc_cut']=test_title_doc['doc'].apply(lambda x:','.join(cut_byPos(x)))

test_title_doc['article_cut']=test_title_doc['title_cut']+test_title_doc['doc_cut']
test_title_doc['article_cut'].to_csv('data/fenci_doc.txt',index=None,header=None)
```
